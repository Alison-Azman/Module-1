```python
#populate a nested list

rows=3
cols=3

# the underscore means we do not need the value and can throw it away

val=0

grid =[]
for _ in range(rows):
  row=[]
  for _ in range (cols):
    row.append(val)
    val+=1
  grid.append(row)

#this creates 3 lists containing empty 3 times all in a list.
print(grid)
```
