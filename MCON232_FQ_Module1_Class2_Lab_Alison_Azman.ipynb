What is a tuple
- a fixed sequence
- immuatble
- lightweight
  - very few operations you can do with them

  When we say tuples are lightweight, we are referring to both their implementation and their intended use.
Fewer built-in operations

- Tuples do not support list methods such as append, insert, or remove. Because they cannot change
size, they require less internal machinery than lists.

- Immutability enables efficiency
Since tuples are immutable, Python can handle them more efficiently in some situations. Tuples are often
faster to create and iterate over than lists, though the performance difference is usually small.

- Conceptually lightweight
Tuples are designed to group together a small, fixed set of related values that are meant to stay together
(for example, coordinates or paired data). They represent structure rather than a collection meant to grow or
change.

- Hashable when contents are hashable
Tuples can be used as dictionary keys or elements of a set. Hashable means Python can compute a stable
(won't change) integer value for the tuple, allowing for fast lookups in hash-based data structures.

When Should You Use a Tuple?

Use a tuple when:

. Values belong together

. The structure should never change

. You want to model a record, not a collection
